```python
# Importing necessary libraries and dependencies
import findspark
findspark.init()
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *

import pandas as pd
import numpy as np
import pymongo
from pymongo import MongoClient

import matplotlib.pyplot as plt
```

```python
# Creating spark session 
spark = SparkSession\
    .builder\
    .master('local[2]')\
    .appName('food_wastage')\
    .config('spark.jars.packages', 'org.mongodb.spark:mongo-spark-connector_2.12:2.4.1')\
    .getOrCreate()
```

```python
# Load csv dataset
db_load = spark.read.csv (r"E:\CS4010-Data Management\Project\restaurant-1-orders.csv", header=True)
db_load2 = spark.read.csv (r"E:\CS4010-Data Management\Project\restaurant-2-orders.csv", header=True)

# Store the dataset into MongoDB
db_load.write.format('mongo').mode('overwrite')\
.option('spark.mongodb.output.uri', 'mongodb://127.0.0.1:27017/food_wastage.restaurant-1-orders').save()
db_load2.write.format('mongo').mode('overwrite')\
.option('spark.mongodb.output.uri', 'mongodb://127.0.0.1:27017/food_wastage.restaurant-2-orders').save()
```

```python
db_load.show(2)
db_load2.show(2)
```

```python
# Separating Date and Time 
s_col = pyspark.sql.functions.split(db_load['Order Date'], ' ')
db_load = db_load.withColumn('Date', s_col.getItem(0))
db_load = db_load.withColumn('Time', s_col.getItem(1))
db_load.show(2)

s_col2 = pyspark.sql.functions.split(db_load2['Order Date'], ' ')
db_load2 = db_load2.withColumn('Date', s_col2.getItem(0))
db_load2 = db_load2.withColumn('Time', s_col2.getItem(1))
db_load2.show(2)
```

```python
# Separating Hour from Time
hour_col = pyspark.sql.functions.split(db_load['Time'], ':')
db_load = db_load.withColumn('Hour', hour_col.getItem(0))
db_load.show(2)

hour_col2 = pyspark.sql.functions.split(db_load2['Time'], ':')
db_load2 = db_load2.withColumn('Hour', hour_col2.getItem(0))
db_load.show(2)
```

```python
# Checking Data format
db_load.printSchema()
db_load2.printSchema()
```

```python
# Separating Day, Month, Year from Date
s_col = pyspark.sql.functions.split(db_load['Date'], '/')
db_load = db_load.withColumn('Day', s_col.getItem(0))
db_load = db_load.withColumn('Month', s_col.getItem(1))  
db_load = db_load.withColumn('Year', s_col.getItem(2))

s_col2 = pyspark.sql.functions.split(db_load2['Date'], '/')
db_load2 = db_load2.withColumn('Day', s_col2.getItem(0))
db_load2 = db_load2.withColumn('Month', s_col2.getItem(1))  
db_load2 = db_load2.withColumn('Year', s_col2.getItem(2))

# Setting Data type for Month and Year
db_load = db_load.withColumn('Day', db_load['Day'].cast(IntegerType()))
db_load = db_load.withColumn('Month', db_load['Month'].cast(IntegerType()))

db_load2 = db_load2.withColumn('Day', db_load2['Day'].cast(IntegerType()))
db_load2 = db_load2.withColumn('Month', db_load2['Month'].cast(IntegerType()))

# Converting all Day and Month into 2-digit format 
db_load = db_load.withColumn('Day', format_string("%02d", col('Day')))
db_load = db_load.withColumn('Month', format_string("%02d", col('Month')))

db_load2 = db_load2.withColumn('Day', format_string("%02d", col('Day')))
db_load2 = db_load2.withColumn('Month', format_string("%02d", col('Month')))

```

```python

# String to Date conversion 
# (https://stackoverflow.com/questions/62602720/string-to-date-migration-from-spark-2-0-to-3-0-gives-fail-to-recognize-eee-mmm)
spark.conf.set("spark.sql.legacy.timeParserPolicy","LEGACY")
db_load = db_load.select("*", concat(col("Year"),lit("-"),col("Month"),lit("-"),col("Day")).alias("New Date"))
db_load = db_load.withColumn("New Date",to_timestamp("New Date"))

db_load2 = db_load2.select("*", concat(col("Year"),lit("-"),col("Month"),lit("-"),col("Day")).alias("New Date"))
db_load2 = db_load2.withColumn("New Date",to_timestamp("New Date"))
```

```python
# Checking Data format
db_load.printSchema()
db_load.show(2)

db_load2.printSchema()
db_load2.show(2)
```

```python
# Finding weekdays for each New_Date
db_load = db_load.withColumn("WeekDay",date_format("New Date","E"))
db_load.show(2)

db_load2 = db_load2.withColumn("WeekDay",date_format("New Date","E"))
db_load2.show(2)
```

```python
# Loading ingredients dataset 
n_db = spark.read.csv (r"E:\CS4010-Data Management\Project\restaurant-1-ingredients.csv", header=True)
n_db.show(2)
n_db2 = spark.read.csv (r"E:\CS4010-Data Management\Project\restaurant-2-ingredients.csv", header=True)
n_db2.show(2)

#Store Dataset into MongoDB
db_load.write.format('mongo').mode('overwrite')\
.option('spark.mongodb.output.uri', 'mongodb://127.0.0.1:27017/food_wastage.restaurant-1-ingredients').save()
db_load2.write.format('mongo').mode('overwrite')\
.option('spark.mongodb.output.uri', 'mongodb://127.0.0.1:27017/food_wastage.restaurant-2-ingredients').save()
```

```python
# Generate Bar Chart for all ingredients 
def plotBar_ingredients(restaurantNo, month, year):
    if restaurantNo == 1:
        if month == '':
            df_temp = restaurant_1_ingredient[(restaurant_1_ingredient["Year"] == year)]
            df1 = df_temp['Ingredients'].value_counts()
        else:
            df_temp = restaurant_1_ingredient[(restaurant_1_ingredient["Year"] == year) & (restaurant_1_ingredient["Month"] == month)]
            df1 = df_temp['Ingredients'].value_counts()
        plt.title("Ingredient Frequency-Restaurant 1"+' [Month: '+str(month)+', Year: '+str(year)+']')
        plt.xlabel("Ingredients")
        plt.ylabel("Amount (unit)")
        df1.plot.bar(figsize = (15,8), color=['#1389FF'])
    elif restaurantNo == 2:   
        if month == '':
            df_temp = restaurant_2_ingredient[(restaurant_2_ingredient["Year"] == year)]
            df1 = df_temp['Ingredients'].value_counts()
        else:
            df_temp = restaurant_2_ingredient[(restaurant_2_ingredient["Year"] == year) & (restaurant_2_ingredient["Month"] == month)]
            df1 = df_temp['Ingredients'].value_counts()
        plt.title("Ingredient Frequency-Restaurant 2"+' [Month: '+str(month)+', Year: '+str(year)+']')
        plt.xlabel("Ingredients")
        plt.ylabel("Amount (unit)")
        df1.plot.bar(figsize = (15,8), color=['#00A591'])
    else:
        print ("Invalid Input")
    
    
   
```

```python
# plotting bar chart (parameters: Restaurant No, Month, Year)
plotBar_ingredients(1,10,2018)
```

```python
# Generating order frequency (2016-2019) for both restaurants
def plotBar_order(restaurantNo, month, year):
    if restaurantNo == 1:
        if month == '':
            df_temp = restaurant_1_order[(restaurant_1_order["Year"] == year)]
            df1 = df_temp['WeekDay'].value_counts()
        else:
            df_temp = restaurant_1_order[(restaurant_1_order["Year"] == year) & (restaurant_1_order["Month"] == month)]
            df1 = df_temp['WeekDay'].value_counts()
        plt.title("Order Frequencies in Weekdays-Restaurant 1")
        plt.xlabel("WeekDays")
        plt.ylabel("Orders")
        df1.plot.bar(color=['#FFA512'])
    elif restaurantNo == 2:
        if month == '':
            df_temp = restaurant_2_order[(restaurant_2_order["Year"] == year)]
            df1 = df_temp['WeekDay'].value_counts()
        else:
            df_temp = restaurant_2_order[(restaurant_2_order["Year"] == year) & (restaurant_2_order["Month"] == month)]
            df1 = df_temp['WeekDay'].value_counts()
        plt.title("Order Frequencies in Weekdays-Restaurant 2")
        plt.xlabel("WeekDays")
        plt.ylabel("Orders")
        df1.plot.bar(color=['#FF6441'])
    else:
        print ("Invalid Input")
    
```

```python
# plotBar_order(restaurantNo, month, year)
plotBar_order(2,6,2019)
```

```python
# plotBar_order(restaurantNo, month, year)
plotBar_order(1,6,2019)
```

```python
# Generating order frequency [pie chart] (2016-2019) for both restaurants 
def plotPie_order(restaurantNo, month, year):
    if restaurantNo == 1:
        df_temp = restaurant_1_order[(restaurant_1_order["Year"] == year) & (restaurant_1_order["Month"] == month)]
        df1 = df_temp['WeekDay'].value_counts()
        df1.plot.pie(figsize = (8,8),)
        plt.legend(bbox_to_anchor = (1.3,0.5),loc = 'right', ncol = 1)
        plt.axis('off')
        plt.title("Restaurant 1 Order Frequency by Month: " + str(month))
    elif restaurantNo == 2:
        df_temp = restaurant_2_order[(restaurant_2_order["Year"] == year) & (restaurant_2_order["Month"] == month)]
        df1 = df_temp['WeekDay'].value_counts()
        df1.plot.pie(figsize = (8,8),)
        plt.legend(bbox_to_anchor = (1.3,0.5),loc = 'right', ncol = 1)
        plt.axis('off')
        plt.title("Restaurant 2 Order Frequency by Month: " + str(month))
    else:
        print ('Invalid Input')
```

```python
# plotPie_order(restaurantNo, month, year)
plotPie_order(1,6,2018)
```

```python
# plotPie_order(restaurantNo, month, year)
plotPie_order(2,2,2018)
```

```python
# Generating mean value of order frequency (2016-2019) for both restaurants
def plotpie_order_mean(restaurantNo):
    if restaurantNo == 1:
        df1 = restaurant_1_order['WeekDay'].value_counts()
        df1.plot.pie(figsize = (8,8),)
        plt.legend(bbox_to_anchor = (1.3,0.5),loc = 'right', ncol = 1)
        plt.axis('off')
        plt.title("Restaurant 1 Order Frequency Mean Value (2016-2019)")
    elif restaurantNo == 2:
        df2 = restaurant_2_order['WeekDay'].value_counts()
        df2.plot.pie(figsize = (8,8),)
        plt.legend(bbox_to_anchor = (1.3,0.5),loc = 'right', ncol = 1)
        plt.axis('off')
        plt.title("Restaurant 2 Order Frequency Mean Value (2016-2019)")
    else:
        print ('Invalid Input')
```
